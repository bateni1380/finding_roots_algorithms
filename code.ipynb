```python
import time
```

```python
# Tamrin 2: Bisection, False Position and Secant

class DebugTools:
    iterations = 0
    
def check_stop_cond(fx, x, last_x, eps, stop_mode):
    if DebugTools.iterations==100: raise Exception('Max iteratin\'s reached!')
    if stop_mode==1:
        return abs(fx)<eps
    if stop_mode==2:
        if last_x==None: return False
        return abs(x-last_x)<eps

# returns root and error
def bisection(f, a, b, eps, stop_mode=2, last_x=None):
    DebugTools.iterations += 1
    if f(a)*f(b)>0:
        raise Exception("The function is not suitable for bisection (f(a)f(b) must be negitive)")
    c = (a+b)/2
    if check_stop_cond(f(c), c, last_x, eps, stop_mode):
        return c, f(c)
    if f(a)*f(c)<0:
        return bisection(f, a, c, eps, stop_mode, last_x=c)
    elif f(c)*f(b)<0: 
        return bisection(f, c, b, eps, stop_mode, last_x=c)

# returns root and error  
def false_position(f, a, b, eps, stop_mode=1, last_x=None):
    DebugTools.iterations += 1
    if f(a)*f(b)>0:
        raise Exception("The function is not suitable for false_position (f(a)f(b) must be negitive)")
    c = (a*f(b)-b*f(a))/(f(b)-f(a))
    if check_stop_cond(f(c), c, last_x, eps, stop_mode):
        return c, f(c)
    if f(a)*f(c)<0:
        return false_position(f, a, c, eps, stop_mode, last_x=c)
    elif f(c)*f(b)<0:
        return false_position(f, c, b, eps, stop_mode, last_x=c)

# returns root and error
def secant(f, x1, x2, eps, stop_mode=1):
    DebugTools.iterations += 1
    x = (x1*f(x2)-x2*f(x1))/(f(x2)-f(x1))
    if check_stop_cond(f(x), x, x1, eps, stop_mode):
        return x, f(x)
    return secant(f, x1, x, eps, stop_mode)

```

```python
start_time = time.time()
DebugTools.iterations = 0
root, error = bisection(lambda x: x**3-x-2, 1, 2, 1e-10, stop_mode=1)
print("The bisection method (with stop condition 1) ran in {} seconds and {} iterations!".format(time.time()-start_time, DebugTools.iterations))
print("Root of f, using bisection method is {} and it\'s error is {}".format(root, error))
```

```python
start_time = time.time()
DebugTools.iterations = 0
root, error = bisection(lambda x: x**3-x-2, 1, 2, 1e-10, stop_mode=2)
print("The bisection method (with stop condition 2) ran in {} seconds and {} iterations!".format(time.time()-start_time, DebugTools.iterations))
print("Root of f, using bisection method is {} and it\'s error is {}".format(root, error))
```

```python
start_time = time.time() 
DebugTools.iterations = 0
root, error = false_position(lambda x: x**3-x-2, 1, 2, 1e-10, stop_mode=1)
print("The false position method (with stop condition 1) ran in {} seconds and {} iterations!".format(time.time()-start_time, DebugTools.iterations))
print("Root of f, using false position method is {} and it\'s error is {}".format(root, error))
```

```python
start_time = time.time() 
DebugTools.iterations = 0
root, error = false_position(lambda x: x**3-x-2, 1, 2, 1e-10, stop_mode=2)
print("The false position method (with stop condition 2) ran in {} seconds and {} iterations!".format(time.time()-start_time, DebugTools.iterations))
print("Root of f, using false position method is {} and it\'s error is {}".format(root, error))
```

```python
start_time = time.time() 
DebugTools.iterations = 0
root, error = secant(lambda x: x**3-x-2, 1, 2, 1e-10, stop_mode=1)
print("The secant method (with stop condition 1) ran in {} seconds and {} iterations!".format(time.time()-start_time, DebugTools.iterations))
print("Root of f, using secant method is {} and it\'s error is {}".format(root, error))
```
